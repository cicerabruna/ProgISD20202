```python
import plotly.graph_objects as go
import pandas as pd
from scipy.signal import find_peaks
```

```python
data = pd.read_csv("dadosFormatados.csv")
```

```python
time_series = data['Average']
indices = find_peaks(time_series, prominence=1, width=6)[0]
# print (indices)

temposPico = (data['Pnt']).loc[indices]
print (temposPico[4:])
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    y=time_series,
    mode='lines+markers',
    name='Plot'
))

fig.add_trace(go.Scatter(
    x=indices,
    y=[time_series[j] for j in indices],
    mode='markers',
    marker=dict(
        size=8,
        color='red',
        symbol='cross'
    ),
    name='Pico'
))

fig.show()
```
